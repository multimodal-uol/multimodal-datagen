```python
from __future__ import absolute_import, division, print_function, unicode_literals
import app
import os,sys
import time
import json, csv
sys.path.append('/opt/anaconda3/lib/python3.7/site-packages')
from clarifai.rest import ClarifaiApp

splunkhome = os.environ['SPLUNK_HOME']
sys.path.append(os.path.join(splunkhome, 'etc', 'apps', 'searchcommands_app', 'lib'))
from splunklib.searchcommands import dispatch, StreamingCommand, Configuration, Option, validators
from splunklib import six
import urllib.request
import requests
from datetime import datetime

jsonImageDir = '/opt/newspaperdata/imagedata/'
jsonVideoDir = '/opt/newspaperdata/videodata/'
actualImageDir = '/opt/images/'
listArticles = []  
class imageObj:
    def __init__(self, title, image,   current_time):
        self.current_time = current_time
        self.image = image
        self.title = title
        
class videoObj:
    def __init__(self, title, movie,   current_time):
        self.current_time = current_time
        self.movie = movie
        self.title = title
        
        
def write_to_json (file_name):
    with open(file_name, 'w') as f:
        if listArticles:
                json.dump([ob.__dict__ for ob in listArticles], f)
                

def detectImage(imageURL):
    app = ClarifaiApp()
    model = app.public_models.general_model
    allConcepts='' 
    app = ClarifaiApp()
    response = model.predict_by_url(imageURL)
    concepts = response['outputs'][0]['data']['concepts']
    for concept in concepts:
        allConcepts=allConcepts+ ' '+concept['name']
    return allConcepts

          
@Configuration()
class NewsImageGen(StreamingCommand):
    totalVideos = Option(
        doc='''
        **Syntax:** ***totalImages=3
        **Description:** Restricts image search''',
        require=True)
    totalImages = Option(
        doc='''
        **Syntax:** ***totalImages=3
        **Description:** Restricts image search''',
        require=True)
    addDescription= Option(
        doc='''
        **Syntax:** ***totalImages=3
        **Description:** Restricts image search''',
        require=False)
    retrieveImage= Option(
        doc='''
        **Syntax:** ***totalImages=3
        **Description:** Restricts image search''',
        require=False)
    retrieveVideo= Option(
        doc='''
        **Syntax:** ***totalImages=3
        **Description:** Restricts image search''',
        require=False)
    def stream(self, records):
        self.logger.debug('CountMatchesCommand: %s', self)  
        image = ''
        if self.totalImages == 'all':
            self.totalImages =100
        if self.totalVideos == 'all':
            self.totalVideos =100
        self.logger.debug('CountMatchesCommand: %s', self)  # logs command line
        image = ''
        timeNow= datetime.now().strftime("%H%M%S%f")
        number_video=0
        
        for record in records:
            current_time = record['current_time']
            if  not os.path.exists(actualImageDir+current_time+'/'):
                os.mkdir(actualImageDir+current_time+'/')
            if  not os.path.exists("/tmp/"+current_time+".txt"):
                os.mknod("/tmp/"+current_time+".txt")
            fileList = os.listdir(actualImageDir+current_time+'/') 
            number_files = len(fileList)
            search_id = current_time
            title = record['Title']
            date_published = record['date_published'][0:19]
            news_outlet = record['news_outlet']
            feature_img = record['feature_img']
            movies = record['movies{}']
            if movies :
                f = open("/tmp/commands.txt", "a")
                f.write(str(type(movies))+"\n")
                f.close()
            article_link = record['article_link']
            if self.retrieveImage =='True':
                if (number_files < int(self.totalImages)):
                    r = requests.get(feature_img, allow_redirects=True)
                    timeNow= datetime.now().strftime("%H%M%S%f")
                    open(actualImageDir+current_time+'/'+timeNow+'.jpg', 'wb').write(r.content)
                    if self.addDescription =='True':
                        image= detectImage(feature_img)
                    anArticle= imageObj (title, image,  current_time)
                    listArticles.append(anArticle)
                    yield{'title':title,'date_published':date_published, 'news_outlet':news_outlet, 
                         'feature_img':feature_img, 'image': image, 'article_link':article_link, '_time':time.time()
                          ,'current_time':current_time}
            elif self.retrieveVideo =='True':
                if (number_video< int(self.totalVideos)):
                    if movies:
                        number_video = sum(1 for line in open("/tmp/"+current_time+".txt"))
                        if number_video == int(self.totalVideos):
                            break
                        elif number_video < int(self.totalVideos):
                            if str(type(movies)) == '<class \'str\'>':
                                videoFile = open("/tmp/"+current_time+".txt", "a")
                                videoFile.write(movies+"\n")
                                videoFile.close()
                                anArticle= videoObj (title, movies,  current_time)
                                listArticles.append(anArticle)
                                yield{'title':title,'date_published':date_published, 'news_outlet':news_outlet, 
                                      'article_link':article_link, '_time':time.time()
                                      ,'current_time':current_time, 'movies':movies}
                            elif str(type(movies)) == '<class \'list\'>':
                                   for movie in movies:
                                        number_video = sum(1 for line in open("/tmp/"+current_time+".txt"))
                                        if number_video < int(self.totalVideos):
                                            videoFile = open("/tmp/"+current_time+".txt", "a")
                                            videoFile.write(movie+"\n")
                                            videoFile.close()
                                            anArticle= videoObj (title, movies,  current_time)
                                            listArticles.append(anArticle)
                                            yield{'title':title,'date_published':date_published, 'news_outlet':news_outlet, 
                                                  'article_link':article_link, '_time':time.time()
                                                  ,'current_time':current_time, 'movies':movie}
                                        else:
                                            break
                                       
                               
                                
            number_video = sum(1 for line in open("/tmp/"+current_time+".txt"))
            if number_video == int(self.totalVideos) or number_files == int(self.totalImages): 
                if self.retrieveImage =='True':
                    f = open("/tmp/commands.txt", "w")
                    f.write(jsonImageDir+current_time+'.json')
                    f.close()
                    if  not os.path.exists(jsonImageDir+current_time+'.json'):
                        write_to_json(jsonImageDir+current_time+'.json')
                        return
                elif self.retrieveVideo =='True':
                    if  not os.path.exists(jsonVideoDir+current_time+'.json'):
                        write_to_json(jsonVideoDir+current_time+'.json')
                        return
        
        


dispatch(NewsImageGen, sys.argv, sys.stdin, sys.stdout, __name__)
```
